Step 3: Apply Named Entity Recognition (NER) Using SpaCy

```python
from google.cloud import bigquery
import pandas as pd
import os

# Set GOOGLE_APPLICATION_CREDENTIALS environment variable
# Replace 'path/to/your/keyfile.json' with the actual path to your service account key file
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "path/to/your/keyfile.json"

# Initialize BigQuery client
client = bigquery.Client()

# Define the BigQuery table
table_id = "plexiform-shine-450303-d9.noteevents.noteeventstbl"

# SQL query to extract discharge summaries
query = f"""
    SELECT SUBJECT_ID, HADM_ID, TEXT
    FROM `{table_id}`
    WHERE CATEGORY = 'Discharge summary'
    LIMIT 100;
"""

# Execute query and load results into Pandas DataFrame
df = client.query(query).to_dataframe()

# Display sample records
print("🔹 Sample Medical Notes Extracted from BigQuery:")
print(df.head())
```

```python
import spacy
# Load pre-trained SpaCy model
nlp_spacy = spacy.load("en_core_web_sm")

# Function to extract entities from text
def extract_spacy_entities(text):
    doc = nlp_spacy(text)
    return [(ent.text, ent.label_) for ent in doc.ents]

# Apply SpaCy NER to medical notes
df["SpaCy_Entities"] = df["TEXT"].apply(lambda x: extract_spacy_entities(str(x)))

# Display extracted entities
print("\n🔹 SpaCy Named Entity Recognition (NER) Results:")
print(df[["SUBJECT_ID", "HADM_ID", "SpaCy_Entities"]].head())

# Save extracted entities to CSV
df.to_csv("medical_notes_with_entities_SpaCy.csv", index=False)
```

Step 5: Train Word2Vec on Medical Notes

```python
from google.cloud import bigquery
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import pandas as pd
import nltk
import os




# Ensure NLTK tokenizer is available
nltk.download('punkt')
nltk.download('punkt_tab')

# Authenticate & Initialize BigQuery Client
os.environ["GOOGLE_CLOUD_PROJECT"] = "plexiform-shine-450303-d9"  # Set project ID
client = bigquery.Client()

# Define BigQuery dataset & table
BQ_PROJECT_ID = "plexiform-shine-450303-d9"  # Google Cloud Project ID
BQ_DATASET_ID = "noteevents"  # Dataset name
BQ_TABLE_ID = "noteeventstbl"  # Table name

# Construct SQL Query to fetch TEXT column from BigQuery
query = f"""
SELECT TEXT FROM `{BQ_PROJECT_ID}.{BQ_DATASET_ID}.{BQ_TABLE_ID}`
WHERE TEXT IS NOT NULL AND LENGTH(TEXT) > 10
LIMIT 10000;
"""

# Run query and load results into Pandas DataFrame
try:
    df = client.query(query).to_dataframe()
    print(f"✅ Data loaded from BigQuery: {df.shape[0]} rows.")
except Exception as e:
    print(f"❌ Error loading data from BigQuery: {e}")
    exit()

# Ensure the "TEXT" column exists and contains valid text
if "TEXT" not in df.columns or df["TEXT"].isnull().all():
    print("❌ No valid TEXT data found in the dataset. Exiting.")
    exit()

# Tokenize and preprocess text for Word2Vec
print("🔄 Tokenizing text for Word2Vec...")
df["TEXT"] = df["TEXT"].astype(str)  # Ensure all text is string
sentences = [word_tokenize(text.lower()) for text in df["TEXT"] if isinstance(text, str)]

# Train Word2Vec model
if sentences:
    print("🔄 Training Word2Vec model...")
    word2vec_model = Word2Vec(sentences, vector_size=100, window=5, min_count=2, workers=4)
    print("✅ Word2Vec model training complete!")

    # Find similar words to "aneurysm"
    try:
        similar_words = word2vec_model.wv.most_similar("aneurysm")
        print("\n🔹 Word2Vec Similar Words for 'aneurysm':")
        for word, similarity in similar_words:
            print(f"{word}: {similarity:.4f}")
    except KeyError:
        print("❌ 'aneurysm' not found in vocabulary. Try increasing dataset size or lowering `min_count`.")
else:
    print("❌ No valid tokenized sentences for Word2Vec training.")



```

Step 6: Apply t-SNE for Word2Vec Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# Extract word vectors from Word2Vec model
word_vectors = np.array([word2vec_model.wv[word] for word in word2vec_model.wv.index_to_key])

# Apply t-SNE for dimensionality reduction
tsne = TSNE(n_components=2, random_state=42)
reduced_vectors = tsne.fit_transform(word_vectors)

# Plot t-SNE visualization
plt.figure(figsize=(10, 5))
plt.scatter(reduced_vectors[:, 0], reduced_vectors[:, 1])

# Annotate words
words = list(word2vec_model.wv.index_to_key)
for i, word in enumerate(words[:10]):  # Show only first 10 words
    plt.annotate(word, (reduced_vectors[i, 0], reduced_vectors[i, 1]))

plt.title("t-SNE Visualization of Word2Vec Medical Terms")
plt.show()
```

```python
#Step 3: Load Medical Notes from BigQuery

from google.cloud import bigquery
import pandas as pd

# Initialize BigQuery client
client = bigquery.Client()

# Define BigQuery table
table_id = "plexiform-shine-450303-d9.noteevents.noteeventstbl"

# SQL query to extract medical notes
query = f"""
    SELECT SUBJECT_ID, HADM_ID, TEXT
    FROM `{table_id}`
    WHERE CATEGORY = 'Discharge summary'
    LIMIT 100;
"""

# Load data into Pandas DataFrame
df = client.query(query).to_dataframe()

# Display sample records
print("🔹 Sample Medical Notes Extracted from BigQuery:")
print(df.head())
```

```python
#Step 4: Extract ClinicalBERT Word Embeddings

def get_bert_embeddings(text):
    """Extracts embeddings from ClinicalBERT for the given text."""
    tokens = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512)
    with torch.no_grad():
        outputs = model(**tokens)
    return outputs.last_hidden_state.mean(dim=1).squeeze().numpy()  # Get sentence embedding

# Apply ClinicalBERT embeddings to medical notes
df["ClinicalBERT_Embeddings"] = df["TEXT"].apply(lambda x: get_bert_embeddings(str(x)))

# Convert embeddings to NumPy array
embeddings = np.vstack(df["ClinicalBERT_Embeddings"].values)

print("✅ Extracted ClinicalBERT embeddings for medical notes!")
```

```python
#Step 5: Apply t-SNE for Dimensionality Reduction

from sklearn.manifold import TSNE

# Apply t-SNE to reduce embeddings to 2D space
tsne = TSNE(n_components=2, random_state=42)
reduced_embeddings = tsne.fit_transform(embeddings)

print("✅ t-SNE transformation applied successfully!")
```

```python
#Step 6: Visualize ClinicalBERT Word Relationships
import matplotlib.pyplot as plt
import seaborn as sns

# Create t-SNE visualization
plt.figure(figsize=(10, 6))
sns.scatterplot(x=reduced_embeddings[:, 0], y=reduced_embeddings[:, 1])

# Annotate a few points with their corresponding text
for i, text in enumerate(df["TEXT"].values[:200]):  # Show only first 200 texts
    plt.annotate(text[:30], (reduced_embeddings[i, 0], reduced_embeddings[i, 1]))

plt.title("t-SNE Visualization of ClinicalBERT Medical Notes")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
#plt.show()


# Save the plot as an image file
plt.savefig("ClinicalBERT_tsne.png", dpi=300, bbox_inches='tight')  # Save at high resolution
plt.show()
```
